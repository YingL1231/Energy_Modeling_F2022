```python
import pandas as pd 
import numpy as np 
egrid=pd.read_excel("egrid2020_data.xlsx",sheet_name="PLNT20",keep_default_na=False,header=1)
egrid.head(3) 
```

## 1. Present a table that shows for each state the number of power plants of different fuel types

We will use the pandas.pivtot_table with syntaxis:
pandas.pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, dropna=True, margins_name='All', observed=False, sort=True)[source]¶

```python
pivot1=pd.pivot_table(egrid,index="PSTATABB",values="PNAME",columns="PLFUELCT",aggfunc='count')
pivot1.head(10)
```

## 2. Present a table that shows for each state the total installed capacity of different fuel types. Leave blank the cells for which there is not a value

```python
egrid.NAMEPCAP=pd.to_numeric(egrid.NAMEPCAP)
pivot2=pd.pivot_table(egrid,index="PSTATABB",values="NAMEPCAP",columns="PLFUELCT",aggfunc='sum',fill_value="")
pivot2.head(10)
```

# 3. Present a table that summarizes for each state info on coal, gas, and nuclear plants, showing the size of the smallest plant, the largest plant, the average size, and the total installed capacity

```python
egrid_sub=egrid[(egrid.PLFUELCT=="COAL")|(egrid.PLFUELCT=="GAS")|(egrid.PLFUELCT=="NUCLEAR")]
pivot3=pd.pivot_table(egrid_sub,index="PSTATABB",values=["NAMEPCAP"],columns="PLFUELCT",aggfunc=[min,max,np.mean,sum],fill_value="")
pivot3.head(10)
```

## 4. What is the average capacity factor and average nameplate capacity of coal, gas and nuclear plants in each state?

```python
egrid_sub.CAPFAC=pd.to_numeric(egrid_sub.CAPFAC)
pivot4=pd.pivot_table(egrid_sub,index="PSTATABB",values=["NAMEPCAP","CAPFAC"],columns="PLFUELCT",aggfunc=[np.mean],fill_value="")
pivot4.tail(10)
```

## 5. Repeat 4, but this time show for each state the total nameplate capacity and the average CF for gas, nuclear, and coal plants

```python
##The code below will give us a wrong table because it will be summing the capacity factors of a given fuel type
## in a state. That is because we are specifying the aggfunc as a list, without saying which function applies to which
## value (i.e., nameplate capacity or capacity factor)
pivot5=pd.pivot_table(egrid_sub,index="PSTATABB",values=["NAMEPCAP","CAPFAC"],columns="PLFUELCT",aggfunc=[sum,np.mean],fill_value="")
pivot5.tail(10)
```

```python
#This code below would give us the desired result. It uses a DICTIONARY to specify that we want to sum the capacity
#but we want to average the capacity factor
pivot6=pd.pivot_table(egrid_sub,index="PSTATABB",values=["NAMEPCAP","CAPFAC"],columns="PLFUELCT",aggfunc={"NAMEPCAP":sum,"CAPFAC":np.mean},fill_value="")
pivot6.tail(10)
```

```python
#works write the markdown
pivot6=pd.pivot_table(egrid_sub,index="PSTATABB",values=["NAMEPCAP","CAPFAC"],columns="PLFUELCT",aggfunc={"NAMEPCAP":sum,"CAPFAC":np.mean},fill_value="",margins=True)
pivot6.tail(10)
```

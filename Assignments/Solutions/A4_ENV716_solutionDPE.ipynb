### 2a (10%) 
Write a function that takes coal price (dollars/ton), coal Heating Value (Btu/lb), and gives coal price in (dollars/MMBtu). 
(The order of the parameters is 1) coal price, and 2) Heating value). 
❗*Remember: there are 2000lbs in 1 ton. MM means million)*


```python
#0. Define 'constants to convert units
LBS_TO_TON = 1/2000 
BTU_TO_MBTU= 1/1e6

#1. Define function
def calculate_coal_price(coal_price_ton, heating_value):
  # Pre: Coal_price_ton is the price of a ton of coal in US dollars (i.e., $/ton)
  #      heating value, is the heat content in a pound of coal (i.e., Btu/lb)
  """ Estimate coal price in Mbtu, using the heating value and price in tons"""
  coal_price_lb = coal_price_ton * LBS_TO_TON  # conversion from tons to lbs
  heating_val_mbtu = heating_value * BTU_TO_MBTU #value is in Btu
  coal_price_mbtu = coal_price_lb / heating_val_mbtu
  return coal_price_mbtu

#2. Define some example inputs for this function (these values are arbitrary and just for demonstration)
price_ton = 100
heat_lb = 7500

#3. Call function and print results
price_mbtu = calculate_coal_price(price_ton, heat_lb)
print("Coal price in $/MBtu: {:,.2f}".format(price_mbtu))
```

### 2b (10%) 
Write a UDF that takes as an input the quantity of any of the greenhouse gases (GHG) reported in eGrid $(CH_4, N_2O,$ and $CO_2)$ in any units (lbs, tons, tonnes), and outputs the $CO_2$ equivalency in metric tonnes . The function should work only if the greenhouse gases are spelled as “CH4”, “N2O” and “CO2”, and the units are spelled as “Lbs”, “Tons”, “Tonnes”. 

```python
#0. Define 'constants to estimate C02 equivalent and unit conversions
# Reference: https://www.epa.gov/energy/greenhouse-gas-equivalencies-calculator
# Releasing 1 kg of CH4 into the atmosphere is about equivalent to 25 kg of CO2
# Releasing 1 kg of N20 into the atmosphere is about equivalent to 298 kg of CO2
CH4_C02_EQ_EMISS = 25
N20_C02_EQ_EMISS = 298
CO2_C02_EQ_EMISS = 1

LBS_TO_TONNE = 1/2205 
TON_TO_TONNE = 2000/2205
TONNES_TO_TONNE = 1.0

#1. Define function
def ghg_co2_equivalent(gas_emiss, gas_name, unit):
  """ Function that takes the quantity of any of the greenhouse gases (GHG) 
  reported in eGrid (CH4, N2O and CO2) in any units (lbs, tons, tonnes), 
  and outputs the CO2 equivalency in metric tonness"""

  # 1st, we estimate the CO2 emissions equivalencies of the input ghg
  if gas_name == 'CH4':
    gas_emiss = gas_emiss * CH4_C02_EQ_EMISS

  elif gas_name == 'N2O':
    gas_emiss = gas_emiss * N20_C02_EQ_EMISS

  elif gas_name == "CO2":
    gas_emiss = gas_emiss * CO2_C02_EQ_EMISS

  else:
   print("Error: The greenhouse gas is not valid or is misspelled")
   gas_emiss = np.NaN
   #Recommedation: use raise function instead of the print and NaN lines aboce. 
   #This will raise an error and stop the excecution of our function
   #raise Exception("Error: The greenhouse gas name is not valid or is misspelled")
   
  # 2nd, we estimate units conversions
  if unit == 'Lbs':
    co2_metric_ton = gas_emiss * LBS_TO_TONNE

  elif unit == 'Tons':
    co2_metric_ton = gas_emiss * TON_TO_TONNE

  elif unit == "Tonnes":
    co2_metric_ton = gas_emiss * TONNES_TO_TONNE

  else:
   print("Error: The greenhouse unit is not valid or is misspelled")
   co2_metric_ton = np.NaN
   #Recommedation: use raise function instead of the print and NaN lines aboce. 
   #This will raise an error and stop the excecution of our function
   #raise Exception("Error: The greenhouse gas unit is not valid or is misspelled")

  return co2_metric_ton

#3. Call function and print results
ghg_emission= 3000
ghg_name = "CO2"
ghg_unit = "Lbs"

co2_equiv_emiss = ghg_co2_equivalent(ghg_emission, ghg_name, ghg_unit)
print("CO2-eq of {:,.2f} {:} {:} is: {:,.2f} metric tonnes".format(ghg_emission, ghg_name, ghg_unit, co2_equiv_emiss))
```

```python
###############################################################################################################
## 1. Read eGRID20 file directly form EPA website
## Read the GEN20 sheet of the excel file into a pandas DataFrame.
###############################################################################################################
df_gen =pd.read_excel("https://www.epa.gov/system/files/documents/2022-01/egrid2020_data.xlsx", 
                      sheet_name="GEN20",keep_default_na=False,header=1)
df_gen.head(n=3) 
```

```python
## 2. Take a subset of the dataframe with only the two columns we care about plus
df_gen_sub=df_gen.loc[:,["PSTATABB","NAMEPCAP","GENYRONL"]]#take all rows but only the two columns
```

```python
## 3. Convert attributes to numeric
df_gen_sub['NAMEPCAP']=pd.to_numeric(df_gen_sub.NAMEPCAP)
df_gen_sub['GENYRONL']=pd.to_numeric(df_gen_sub.GENYRONL) 
```

```python
#Optional: the conversion to numeric, cconverted both columns to float
#we can convert the years into integers
df_gen_sub.GENYRONL = df_gen_sub['GENYRONL'].astype('Int64', errors='ignore')
```

```python
## 4. Get a subset of the dataframe with only the generators built in the period 2010 to 2020
df_gen_sub2 = df_gen_sub[(df_gen_sub.GENYRONL >= 2010) & (df_gen_sub.GENYRONL <=2020)].copy() #We will explain copy() later
```

```python
## 5. Create a new column in the df_gen_sub2 whose values can be counted in a pivot table we will create in a bit.
## The rows of the pivot table will be the states, the columns will be years online from 2010 to 2020
## And the values of the pivot table? something we can count, to count the number of generators in each state, 
## that came online in each year.
## You could just create a dummy column that repeats the GENYRONL and count those values. Or we can create any other column.
## The line below creates a new coulumn with ranks of the generators by year online, from youngest to oldest
## it is not the rank we need in the last step (because we need to rank states by number of gens) but we practice using rank
## and we get a column that we can use as "values"
df_gen_sub2['RANK_GENS'] = df_gen_sub2['GENYRONL'].rank(method='min', ascending=False)
## and alternative would be to create a dummy column that has a copy of the GENYRONL values:
```

```python
df_gen_sub2
```

```python
#4. Create pivot table with the count of generators that were online from 2010 to 2020 by state
# Get to total number of gens installed in a state by adding the parameters margins=True, margins_name='Total'
pt_online_2010_2020 = pd.pivot_table(df_gen_sub2,index="PSTATABB", columns='GENYRONL', 
                                     values='RANK_GENS', aggfunc='count', fill_value='', 
                                     margins = True, margins_name="TOTAL_GENS_ONLINE_2010_2020")
```

```python
## We see there is a column for Puerto Rico (which is not on the list of the 50+1 states)
## We can drop it, but no points are taken off for not doing this
pt_online_2010_2020.drop(index=('PR'), inplace=True) #Having "inplace=True" means that we do not create a new dataframe, but just modify the one we are working on
```

```python
## There is also an extra row with the total of gens online in each year. We do not need it so we drop it
pt_online_2010_2020.drop(index=('TOTAL_GENS_ONLINE_2010_2020'), inplace=True)
```

```python
# 5. Rank each state by the total number of generators installed, in descending order, so the state with most generators is 1
pt_online_2010_2020['RANK_GENS_ONLINE_2010_2020'] = pt_online_2010_2020['TOTAL_GENS_ONLINE_2010_2020'].rank(method='min', ascending=False)
pt_online_2010_2020.head(n=3)
```

```python
#6 Calculate total capacity installed using a second pivot table with margins
# Add totals to the pivot table with the margin
pt_total_cap = pd.pivot_table(df_gen_sub2,index="PSTATABB", values='NAMEPCAP', 
                              aggfunc='sum', fill_value='')

#6.1 Drop data from Puerto Rico
pt_total_cap.drop(index=('PR'), inplace=True)

pt_total_cap.head(n=5)
```
